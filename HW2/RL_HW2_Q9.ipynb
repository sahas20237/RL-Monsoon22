```python
import numpy as np
```

```python
grid = np.zeros((4, 4)) #initializing grid of 4x4
# temp grid is needed to store the value of v during one iteration v calcuation
# after each episode, the final value is stored in the main grid at the end
grid_temp = np.zeros((4, 4))
gamma = 1 #discounting factor (no discounting)
p = 0.25
r = -1
```

```python
for _ in range(120): #120 eps ~ infinity
    for i in range(4): #iterating over grid
        for j in range(4):
            # handling the special case (left top corner and right bottom corner)
            if (i == 0 and j == 0) or (i == 3 and j == 3):
                continue
            else:
                up = None
                down = None
                right = None
                left = None

                #this is a flag that saves us from going out of the grid
                if i > 0:
                    up = grid_temp[i-1][j]
                if i < 3:
                    down = grid_temp[i+1][j]
                if j > 0:
                    left = grid_temp[i][j-1]
                if j < 3:
                    right = grid_temp[i][j+1]

                # v = reward + Y*v[next state]
                v = 0
                if up != None:
                    v += p*(r + gamma*up)
                else:
                    v += p*(r + gamma*grid_temp[i][j])

                if down != None:
                    v += p*(r + gamma*down)
                else:
                    v += p*(r + gamma*grid_temp[i][j])

                if left != None:
                    v += p*(r + gamma*left)
                else:
                    v += p*(r + gamma*grid_temp[i][j])

                if right != None:
                    v += p*(r + gamma*right)
                else:
                    v += p*(r + gamma*grid_temp[i][j])

                # here we put the final value of v calculated in this iteration 
                # and store it in the main grid for use in the next iteration
                grid[i][j] = v
    
    # here we update the temp grid so that the next iteration 
    # uses the final values of previous iteration
    grid_temp = grid.copy()
    
    # printing all iterations
    grid = np.round(grid, 1)
    print(grid)
```

```python
grid = np.round(grid, 1)
print(grid)
```
